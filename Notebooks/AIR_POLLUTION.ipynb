```python
new_data=data2022[['Time','Kolkata']]
```

```python
# Select 'Time' and 'Kolkata' columns from data2023
new_observations = data2023[['Time', 'Kolkata']]

# Append these new observations to the existing 'data' DataFrame
new_data = pd.concat([new_data, new_observations], ignore_index=True)

print("Updated data DataFrame head:")
display(new_data.head())

print("\nUpdated data DataFrame tail:")
display(new_data.tail())
```

```python
# Select 'Time' and 'Kolkata' columns from data2024
new_observations1 = data2024[['Time', 'Kolkata']]

# Append these new observations to the existing 'data' DataFrame
new_data = pd.concat([new_data, new_observations1], ignore_index=True)

print("Updated data DataFrame head:")
display(new_data.head())

print("\nUpdated data DataFrame tail:")
display(new_data.tail())
```

```python
# Convert 'Time' column to datetime objects, specifying dayfirst=True
new_data['Time'] = pd.to_datetime(new_data['Time'], dayfirst=True)

# Sort data by time to ensure proper plotting
new_data = new_data.sort_values(by='Time')

plt.figure(figsize=(15, 7))
sns.lineplot(x='Time', y='Kolkata', data=new_data)
plt.title('Kolkata PM2.5 level Over Time (2022-2024)')
plt.xlabel('Date')
plt.ylabel('Kolkata PM2.5')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Set 'Time' column as index for seasonal decomposition
df_indexed = new_data.set_index('Time')

# Perform seasonal decomposition (assuming additive model and yearly seasonality)
# Adjust model='multiplicative' if data exhibits increasing or decreasing seasonality amplitude
# Adjust period if seasonality is different (e.g., weekly: 7, monthly: 30 or 31)
decomposition = seasonal_decompose(df_indexed['Kolkata'], model='additive', period=365)

# Plot the decomposed components
fig = decomposition.plot()
fig.set_size_inches(12, 8)
fig.suptitle('Seasonal Decomposition of Kolkata AQI', y=1.02) # Add title and adjust its position
plt.tight_layout(rect=[0, 0, 1, 0.98]) # Adjust layout to prevent title overlap
plt.show()
```

```python
# Calculate a rolling mean to visualize the trend
# Using a 30-day rolling window as an example
new_data['Kolkata_Rolling_Mean'] = new_data['Kolkata'].rolling(window=30).mean()

plt.figure(figsize=(15, 7))
sns.lineplot(x='Time', y='Kolkata', data=new_data, label='Original AQI')
sns.lineplot(x='Time', y='Kolkata_Rolling_Mean', data=new_data, color='red', label='30-Day Rolling Mean')
plt.title('Kolkata AQI with 30-Day Rolling Mean Trend')
plt.xlabel('Date')
plt.ylabel('Kolkata AQI')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

The red line in the plot above represents the 30-day rolling mean, which helps to smooth out daily variations and can sometimes reveal a subtle trend that might not be immediately obvious in the seasonal decomposition's trend component. If this rolling mean also appears relatively flat, it further supports the idea of a stable AQI over this period.
